```python
from roboflow import Roboflow
from IPython.display import Image
rf = Roboflow(api_key="rTeP0X0zEGKBqh9Q3VWn")
project = rf.workspace().project("human-sign-language-translator")
model = project.version(2).model

# infer on a local image
print(model.predict("a1.jpeg", confidence=30, overlap=30).json())

# visualize your prediction
model.predict("a1.jpeg", confidence=30, overlap=30).save("prediction.jpg")
Image("prediction.jpg")

# infer on an image hosted elsewhere
# print(model.predict("URL_OF_YOUR_IMAGE", hosted=True, confidence=40, overlap=30).json())
```
